Bu kod, Python'daki `warnings` modülünü kullanarak belirli bir türdeki uyarıları görmezden gelmek veya filtrelemek için kullanılır. Burada, `FutureWarning` kategorisindeki uyarıları görmezden gelmek istendiği belirtilmiştir.

Bu tür bir kod genellikle, bir kütüphanenin gelecekteki sürümlerinde kullanım dışı (deprecated) olacak bazı özelliklere veya değişikliklere işaret eden uyarıları engellemek veya gizlemek için kullanılır. Ancak, uyarıları görmezden gelmek, gelecekte bir güncelleme veya değişiklik nedeniyle oluşabilecek sorunları önceden tespit etme şansını da azaltabilir.

Eğer bir kütüphanenin gelecekteki sürümleri için uyarıları görmezden gelmek istiyorsanız, bunu yapmadan önce iyi bir nedeniniz olmalıdır. Genellikle, uyarılar, bir kütüphanenin daha güncel sürümlerine geçiş yaparken kodunuzun uyumluluğunu kontrol etmenize ve gerekli güncellemeleri yapmanıza yardımcı olur.

Eğer bu uyarıları filtreleme ihtiyacınız varsa, ancak ne tür uyarılar olduğunu bilmek istiyorsanız, filtrelenmiş uyarılar hakkında bilgi almak için şu kodu kullanabilirsiniz:

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# İşlemleriniz
# ...

# Filtrelenmiş uyarıları gösterme
with warnings.catch_warnings():
    warnings.simplefilter("always", category=FutureWarning)
    # Gelecekteki sürüm uyarıları
    # ...
```

Bu şekilde, belirli bir bölümde filtrelenmiş uyarıları tekrar görebilir ve bu uyarılar hakkında bilgi alabilirsiniz.

### Prosedür 2 - Özellik Mühendisliği

**Adım_1:** Daha iyi analiz için bazı özelliklerin değiştirlmesi veya optimize edilmesi gerekir. Aile büyüklüğünü görmek için **`SibSp`** ve **`Parch`** özelliklerini birleştirebiliriz, çünkü SibS gemideki kardeş/eş sayısını tutarken, Parch gemideki ebeveyn/çocuk sayısı ile ilgilidir.

Bu kod, "SibSp" (kardeş/eş sayısı) ve "Parch" (ebeveyn/çocuk sayısı) sütunlarını kullanarak yeni bir "FamilySize" (aile büyüklüğü) sütunu oluşturur ve bu sütundaki değerlerin sayısını hesaplar.

1. "SibSp" ve "Parch" sütunlarını toplayarak her bir yolcu için aile büyüklüğünü temsil eden yeni bir "FamilySize" sütunu oluşturur.
2. Oluşturulan "FamilySize" sütunundaki her bir değerin sayısını **`value_counts()`** fonksiyonuyla hesaplar.
3. Hesaplanan değerleri ekrana yazdırır.

Bu şekilde, aile büyüklüklerinin farklı değerlerde kaç kez olduğunu görebilirsiniz. Bu, veri setindeki aile büyüklüğü dağılımını anlamak için kullanışlı bir bilgidir.

````python
# "SibSp" ve "Parch" sütunlarını kullanarak "FamilySize" sütununu oluşturma
titanic["FamilySize"]=titanic["SibSp"]+titanic["Parch"]

# "FamilySize" sütunundaki değerlerin sayısını hesaplama ve yazdırma
print(titanic["FamilySize"].value_counts())
````

**Adım_2:** Aile büyüklüğünü dört gruba ayırın; tek kişi, sadece eş, küçük aile (5'ten az) ve büyük aile (5 veya daha fazla aile üyesi).

Bu kod, "FamilySize" sütununu baz alarak yeni bir "FsizeD" (aile büyüklüğü kategorisi) sütunu oluşturur. Bu sütun, "singleton" (yalnız), "spouse" (eş), "small" (küçük) ve "large" (büyük) olmak üzere dört kategoriye ayrılır. Ardından, oluşturulan kategorilerin benzersiz değerlerini ve her bir kategorinin veri setinde kaç kez bulunduğunu gösteren sayıları ekrana yazdırır.

Bu kod, "FamilySize" sütunundaki değerlere göre "FsizeD" sütununu oluşturur ve aile büyüklüğüne dayalı dört farklı kategoriyi belirler. **`unique()`** fonksiyonu, "FsizeD" sütunundaki benzersiz değerleri getirirken, `value_counts()` fonksiyonu ise her bir kategorinin veri setinde kaç kez bulunduğunu gösterir. Bu bilgiler, aile büyüklüğü kategorilerinin dağılımını anlamanıza yardımcı olacaktır.

````python
# "FamilySize" sütununu baz alarak "FsizeD" sütununu oluşturma
titanic.loc[titanic["FamilySize"]==0, "FsizeD"]='singleton'
titanic.loc[titanic["FamilySize"]==1, "FsizeD"]='spouse'
titanic.loc[(titanic["FamilySize"]>1) &
            (titanic["FamilySize"]<5), "FsizeD"]='small'
titanic.loc[titanic["FamilySize"]>4, "FsizeD"]='large'

# "FsizeD" sütunundaki benzersiz değerleri yazdırma
print(titanic["FsizeD"].unique())

# "FsizeD" sütunundaki her bir kategorinin sayısını yazdırma
print(titanic["FsizeD"].value_counts())
````

**Adım_3:** Kategorik değişkenleri sayısal değişkenlere dönüştürmenin bir örneği olarak, Embarked özelliğini kategorik bir özelliğe dönüştürebiliriz, böylece bu kategori değerlerini etiket kodlamamız için kullanabiliriz.

Bu kod, "Embarked" sütununu kategorik bir değişkene dönüştürür ve ardından bu kategorik değişkenin kodlarını (integer olarak ifade edilen kategori kodları) içeren "Embarked" sütununu oluşturur.

1. **`astype('category')`**: "Embarked" sütununu kategorik bir değişkene dönüştürür. Kategorik değişkenler, sınırlı ve belirli değer kümesine sahip olan değişkenlerdir.
2. **`cat.codes`**: Kategorik değişkenin kodlarını (integer olarak ifade edilen kategori kodları) içeren yeni bir "Embarked" sütunu oluşturur.

Bu işlemlerin sonucunda, "Embarked" sütunu artık kategorik değil, kategorik değişkenin kodlarını içeren bir sayısal sütun olacaktır. Bu tür bir dönüşüm, özellikle makine öğrenimi modellerine giriş verilerini hazırlarken kullanılır, çünkü çoğu model sayısal verilerle daha iyi çalışır. Ancak, bu tür bir dönüşüm yaparken dikkatli olmalısınız ve kategorik değişkenin sıralı mı yoksa sırasız mı olduğunu anlamalısınız, çünkü kategorik değişkenlere atanan kodlar bu sıralamayı temsil eder.

````python
# "Embarked" sütununu kategorik bir değişkene dönüştürme
titanic["Embarked"]=titanic["Embarked"].astype('category')

# Kategorik değişkenin kodlarını içeren "Embarked" sütununu oluşturma
titanic["Embarked"]=titanic["Embarked"].cat.codes

# Veri setinin ilk beş gözlemine bakma
titanic.head()
````

**Adım_4:** **`corr()`** yöntemi tüm özellikler arasındaki korelasyonu değerlendirir, daha sonra bir renk kodlaması ile grafik haline getirilebilir.

Bu kod, DataFrame içinde sayısal verilere sahip sütunları seçer ve ardından bu sayısal sütunlar arasındaki korelasyon matrisini oluşturur. Korelasyon matrisi, her bir sayısal değişkenin diğerleri ile olan ilişkisini gösterir.

Bu kodun sonucunda, "data" DataFrame'inin sayısal sütunları arasındaki korelasyonlar içeren bir korelasyon matrisi elde edersiniz. Korelasyon matrisinde, her bir çift sayısal değişken arasındaki korelasyon katsayısı bulunur. Korelasyon katsayısı, -1 ile 1 arasında bir değer alır; -1'e yaklaştıkça negatif korelasyon, 1'e yaklaştıkça pozitif korelasyon, 0'a yaklaştıkça ise ilişki zayıflar.

Bu matris, veri setinizdeki sayısal değişkenler arasındaki ilişkileri anlamak ve çoklu değişken analizi yapmak için kullanışlıdır.

````python
# Sayısal verilere sahip sütunları seçme ve korelasyon matrisini oluşturma
corr=data.select_dtypes(include=np.number).corr()

# Korelasyon matrisini yazdırma
corr
````

Korelasyon matrisi oluşturmanın bir başka yolu da aşağıdaki gibidir;

````python
import pandas as pd

# Veri çerçevenizi yükleyin veya oluşturun
# Örneğin:
# data = pd.read_csv('veri.csv')

# Sayısal olmayan sütunları çıkarın
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Korelasyon matrisini oluşturun
corr_matrix = numerical_data.corr()

# Korelasyon matrisini yazdırın
print(corr_matrix)
````

**Adım_5:** Korelasyon yoğunluklarını göstermek için bir ısı haritası oluşturun. Eksenleri etiketleyin ve bir renk paleti oluşturun.

Bu Python kodu, bir veri çerçevesinden sayısal verileri seçer, bu sayısal veriler arasındaki korelasyon matrisini oluşturur ve ardından bu korelasyon matrisini bir ısı haritası olarak görselleştirir.

Bu kodun temel adımları:

1. **`data.select_dtypes(include=['float64', 'int64'])`**: Bu adım, veri çerçevesinden sadece sayısal (**`float64`** ve **`int64`**) sütunları seçer.

2. **`numerical_data.corr()`**: Bu adım, seçilen sayısal veriler arasında korelasyon matrisini hesaplar.

3. **`plt.matshow`**: Bu adım, ısı haritasını oluşturur ve gösterir.

4. **`fig.colorbar(cax)`**: Bu adım, renk skalasını gösteren bir renk çubuğu ekler.

5. **`ax.set_xticks`, `ax.set_yticks`, `ax.set_xticklabels`, `ax.set_yticklabels`**: Bu adımlar, x ve y eksenlerine etiketleri ekler ve dönerleri ayarlar.

6. **`plt.show()`**: Bu adım, oluşturulan ısı haritasını görüntüler.

Sonuç olarak, bu kod, veri çerçevesindeki sayısal sütunlar arasındaki korelasyonları anlamak için bir ısı haritası sağlar. Isı haritası, renk tonları aracılığıyla korelasyonun gücünü görsel olarak temsil eder.

````python

# Isı haritasını oluşturun
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_matrix, cmap="coolwarm", vmin=-1, vmax=1)
fig.colorbar(cax)

ticks = np.arange(0, len(data.columns), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(data.columns, rotation=90)
ax.set_yticklabels(data.columns)

plt.show()
````
