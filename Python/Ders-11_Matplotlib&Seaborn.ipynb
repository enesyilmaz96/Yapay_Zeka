## MATPLOTLIB

Matplotlib, bilimsel grafikler, çizimler, histogramlar, scatter plot'lar ve diğer çeşitli grafiklerin oluşturulması için geniş bir olanak sunar. Matplotlib, özellikle veri analizi, bilimsel araştırmalar, mühendislik uygulamaları ve öğrenci projeleri gibi birçok alanda yaygın olarak kullanılmaktadır.

Matplotlib'in ana özellikleri şunlardır:

1. **Çeşitli Grafik Türleri:** Matplotlib, çizgi grafikleri, scatter plot'lar, bar grafikleri, histogramlar, pasta grafikleri, yüzey plot'lar ve daha pek çok grafik türü oluşturmak için fonksiyonlar içerir.

2. **Özelleştirilebilirlik:** Grafiklerin görünümü, renkleri, etiketleri ve diğer özellikleri geniş ölçüde özelleştirilebilir. Bu, kullanıcıların grafikleri istedikleri gibi düzenlemelerine olanak tanır.

3. **Matematiksel İşlemler:** Matplotlib, matematiksel ifadelerin, eğrilerin ve fonksiyonların grafiklerini çizme yeteneğine sahiptir.

4. **Interaktif Grafikler:** Jupyter Notebook veya IPython gibi interaktif ortamlarda kullanılabilir ve kullanıcıların grafiklerle etkileşimde bulunmalarına olanak tanır.

5. **Destek ve Topluluk:** Matplotlib, geniş bir kullanıcı topluluğuna sahiptir ve bu topluluk, kullanıcıların sorularına cevap vermek, önerilerde bulunmak ve paylaşılan kaynaklar aracılığıyla yardımcı olmak için aktiftir.


## Kategorik Değişken Görselleştirme

Kategorik değişkenlerde **sütun grafiği** kulanılır. Bu kod bloğu, Titanic veri setindeki cinsiyet (sex) sütunundaki değerlerin sayısını bar grafik olarak göstermektedir. İşte kodun adım adım açıklamaları:

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Pandas'ın tüm sütunları gösterme ve genişlik seçeneğini ayarlama
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

# Seaborn kütüphanesinden "titanic" veri setini yükleme
df = sns.load_dataset("titanic")

# DataFrame'in ilk beş satırını gösterme
df.head()
````

Bunlardan birtanesi örnek olarak gösterelim;

````python
# Cinsiyet (sex) sütunundaki değerlerin sayısını hesaplayıp, 
# kind='bar' kodu ile sütun grafiği çizdireceğimizi belirttik.

df['sex'].value_counts().plot(kind='bar')

# Grafik gösterme
plt.show()
````

## Sayısal Değişken Görselleştirme

Sayısal değişkenlerde **Histogram grafiği** ve **kutu grafiği** kullanılır. Bu kod bloğu, Titanic veri setindeki "age" sütunundaki yaş değerlerini bir histogram (ilk grafik) ve "fare" sütunundaki bilet fiyatlarını bir kutu grafiği (boxplot - ikinci grafik) ile göstermektedir. 

````python
# Pandas'ın tüm sütunları gösterme ve genişlik seçeneğini ayarlama
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

# Seaborn kütüphanesinden "titanic" veri setini yükleme
df = sns.load_dataset("titanic")

# DataFrame'in ilk beş satırını gösterme
df.head()
````

````python
# "age" sütunundaki yaş değerlerini histogram olarak gösterme
plt.hist(df["age"])
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
````

````python
# "fare" sütunundaki bilet fiyatlarını kutu grafiği olarak gösterme
plt.boxplot(df["fare"])
plt.title("Boxplot of Fare")
plt.ylabel("Fare")
plt.show()
````

## plot

Bu kod bloğu, Matplotlib kütüphanesini kullanarak basit bir çizgi grafiği ve scatter plot (nokta grafiği) oluşturmayı içerir.

````python
import numpy as np
import matplotlib.pyplot as plt

# İlk çizgi grafiği
x1 = np.array([1, 8])
y1 = np.array([0, 150])

plt.plot(x1, y1)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
````

````python
# İlk scatter plot (nokta grafiği)
plt.plot(x1, y1, 'o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

````python
# İkinci çizgi grafiği
x2 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])

plt.plot(x2, y2)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

````python
# İkinci scatter plot (nokta grafiği)
plt.plot(x2, y2, 'o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

## line

Bu kod, Matplotlib kütüphanesini kullanarak bir çizgi grafiği oluşturmayı içerir.

````python
# Veri setini oluştur
y = np.array([13, 28, 11, 100])

# Çizgi grafiğini oluştur ve göster
# color ile renginin kırmızı olması gerektiğini
# linestyle ile ise çizgi şeklini belirledik
plt.plot(y, linestyle="dashdot", color="r")

# Grafiğe başlık ekledik
plt.title('Line Plot with Dashdot Line Style and Red Color')
# X eksenine başlık ekledik
plt.xlabel('Index')
# Y eksenine başlık ekledik
plt.ylabel('Values')
plt.show()
````

## Multiple Lines

Bu kod, Matplotlib kütüphanesini kullanarak iki farklı veri setini içeren iki çizgi grafiğini aynı grafik üzerinde çizmeyi içerir.

````python
# İki farklı veri setini oluştur
x = np.array([23, 18, 31, 10])
y = np.array([13, 28, 11, 100])

# İlk veri setini çizgi grafiği olarak ekle
plt.plot(x, label='Line X')
# İkinci veri setini çizgi grafiği olarak ekle
plt.plot(y, label='Line Y')

# Grafiği göster

# Legend eklenerek her bir çizgi grafiğinin neyi temsil ettiği belirtilir
plt.legend() 
plt.title('Multiple Line Plots')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
````

## Subplots

Bu kod, Matplotlib kütüphanesini kullanarak subplotlar içeren bir figür oluşturmayı içerir. İki farklı subplot düzenini içerir. İlk olarak, iki subplot'u bir satır ve iki sütun içinde düzenler, ardından üç subplot'u bir satır ve üç sütun içinde düzenler. 

**İki subplot (1 satır, 2 sütun):**

````python
# Plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# burada 1 satır 2 sütunluk grafik oluşturacağımızı ve bunun ilk grafik 
# olduğu bilgisini verdik.
plt.subplot(1, 2, 1)  # 1 satır, 2 sütun, 1. subplot
plt.title("1")
plt.plot(x, y)

# Plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])

# burada 1 satır 2 sütunluk grafik oluşturacağımızı ve bunun ikinci
# grafik olduğu bilgisini verdik.
plt.subplot(1, 2, 2)  # 1 satır, 2 sütun, 2. subplot
plt.title("2")
plt.plot(x, y)

plt.show()
````

**Üç subplot (1 satır, 3 sütun):**

````python
# Plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 1)  # 1 satır, 3 sütun, 1. subplot
plt.title("1")
plt.plot(x, y)

# Plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 3, 2)  # 1 satır, 3 sütun, 2. subplot
plt.title("2")
plt.plot(x, y)

# Plot 3
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 3)  # 1 satır, 3 sütun, 3. subplot
plt.title("3")
plt.plot(x, y)

plt.show()
````

Her iki durumda da, **plt.subplot** fonksiyonuyla subplotlar belirlenir. İlk iki parametre, subplot düzenini belirler (satır ve sütun sayısı), üçüncü parametre ise hangi subplot'a çizim yapılacağını belirler. Her subplot çizildikten sonra **plt.title** ile başlık eklenir ve **plt.show** ile grafik gösterilir.

## SEABORN

Veri görselleştirme için kullanılan yüksek seviye bir kütüphanedir. Matplotlib ile kıyaslandığında daha az çaba ile grafik çizilebilmektedir.

````python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# Seaborn kütüphanesinde bulunan "tips" veri setini yükle
df = sns.load_dataset("tips")

# "sex" değişkeninin değerlerinin sayısını hesapla ve ekrana yazdır
print(df["sex"].value_counts())
````

````python
# seaborn'un countplot fonksiyonu ile "sex" değişkeninin sayısını görselleştir
sns.countplot(x=df["sex"], data=df)
plt.title("Countplot of 'sex' in the Tips Dataset")  # Grafik başlığı ekleyerek
plt.xlabel("Sex")  # X ekseni etiketi ekleyerek
plt.ylabel("Count")  # Y ekseni etiketi ekleyerek
plt.show()
````

````python
# Pandas'ın plot fonksiyonu ile "sex" değişkeninin sayısını bar grafiği olarak görselleştir
df['sex'].value_counts().plot(kind='bar')
plt.title("Bar Chart of 'sex' in the Tips Dataset")  # Grafik başlığı ekleyerek
plt.xlabel("Sex")  # X ekseni etiketi ekleyerek
plt.ylabel("Count")  # Y ekseni etiketi ekleyerek
plt.show()
````

## Sayısal Değişken Görselleştirme

Bu iki görselleştirme yöntemi, "total_bill" değişkeninin veri setindeki dağılımını farklı şekillerde ifade eder. Boxplot, veri setinin merkezi eğilimi ve yayılımı hakkında genel bilgiler sunarken, histogram ise verinin frekans dağılımını daha ayrıntılı bir şekilde gösterir.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# "total_bill" değişkeninin kutu grafiğini çiz
sns.boxplot(x=df["total_bill"])
plt.title("Boxplot of 'total_bill' in the Tips Dataset")  # Grafik başlığı
plt.xlabel("Total Bill")  # X ekseni etiketi
plt.show()
````

````python
import matplotlib.pyplot as plt

# "total_bill" değişkeninin histogramını çiz
df["total_bill"].hist()
plt.title("Histogram of 'total_bill' in the Tips Dataset")  # Grafik başlığı
plt.xlabel("Total Bill")  # X ekseni etiketi
plt.ylabel("Frequency")  # Y ekseni etiketi
plt.show()
````
