```python
##################################################
# List Comprehensions
##################################################

# ###############################################
# # GÖREV 1: List Comprehension yapısı kullanarak car_crashes verisindeki numeric değişkenlerin 
# isimlerini büyük harfe çeviriniz ve başına NUM ekleyiniz.
# ###############################################
#
# # Beklenen Çıktı
#
# # ['NUM_TOTAL',
# #  'NUM_SPEEDING',
# #  'NUM_ALCOHOL',
# #  'NUM_NOT_DISTRACTED',
# #  'NUM_NO_PREVIOUS',
# #  'NUM_INS_PREMIUM',
# #  'NUM_INS_LOSSES',
# #  'ABBREV']
#
# # Notlar:
# # Numerik olmayanların da isimleri büyümeli.
# # Tek bir list comp yapısı ile yapılmalı.
```

```python
import seaborn as sns
import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.width',500)

df=sns.load_dataset("car_crashes")
df.columns
```

```python
df.info()
```

```python
["NUM_"+col.upper() if df[col].dtype != "O" else col.upper() for col in df.columns]
```

```python
# ###############################################
# # GÖREV 2: List Comprehension yapısı kullanarak car_crashes verisindeki isminde "no" barındırmayan 
# değişkenlerin isimlerininin sonuna "FLAG" yazınız.
# ###############################################
#
# # Notlar:
# # Tüm değişken isimleri büyük olmalı.
# # Tek bir list comp ile yapılmalı.
#
# # Beklenen çıktı:
#
# # ['TOTAL_FLAG',
# #  'SPEEDING_FLAG',
# #  'ALCOHOL_FLAG',
# #  'NOT_DISTRACTED',
# #  'NO_PREVIOUS',
# #  'INS_PREMIUM_FLAG',
# #  'INS_LOSSES_FLAG',
# #  'ABBREV_FLAG']
```

```python
[col.upper() + "_FLAG" if "no" not in col else col.upper() for col in df.columns]
```

```python
# ###############################################
# # Görev 3: List Comprehension yapısı kullanarak aşağıda verilen değişken isimlerinden 
# FARKLI olan değişkenlerin isimlerini seçiniz ve yeni bir dataframe oluşturunuz.
# ###############################################
#
#   og_list = ["abbrev", "no_previous"]
#
# # Notlar:
# # Önce yukarıdaki listeye göre list comprehension kullanarak new_cols adında yeni liste oluşturunuz.
# # Sonra df[new_cols] ile bu değişkenleri seçerek yeni bir df oluşturunuz adını new_df olarak isimlendiriniz.
#
# # Beklenen çıktı:
#
# #    total  speeding  alcohol  not_distracted  ins_premium  ins_losses
# # 0 18.800     7.332    5.640          18.048      784.550     145.080
# # 1 18.100     7.421    4.525          16.290     1053.480     133.930
# # 2 18.600     6.510    5.208          15.624      899.470     110.350
# # 3 22.400     4.032    5.824          21.056      827.340     142.390
# # 4 12.000     4.200    3.360          10.920      878.410     165.630
#
```

```python
og_list=["abbrev","no_previous"]
new_cols=[col for col in df.columns if col not in og_list]
new_df=df[new_cols]
new_df.head()
```

```python
#########################################
# Görev 11: Her bir değişkendeki boş değerlerin toplamını bulunuz.
#########################################

df.isnull().sum()
```

```python
#########################################
# Görev 12: who değişkenini dataframe'den düşürün.
#########################################

df.drop("who", axis=1, inplace=True)
df.head()
```

```python
#########################################
# Görev 13: deck değikenindeki boş değerleri deck değişkenin en çok tekrar eden değeri (mode) ile doldurunuz.
#########################################

type(df["deck"].mode())
```

```python
df["deck"].mode()[0]
```

```python
df["deck"].fillna(df["deck"].mode()[0], inplace=True)
df["deck"].isnull().sum()
```

```python
#########################################
# Görev 14: age değikenindeki boş değerleri age değişkenin medyanı ile doldurun.
#########################################

type(df["age"].median())
```

```python
df["age"].median()
```

```python
df["age"].fillna(df["age"].median(), inplace=True)
df.isnull().sum()
```

```python
#########################################
# Görev 15: survived değişkeninin Pclass ve Cinsiyet değişkenleri kırılımınında sum, count, mean değerlerini bulunuz.
#########################################

df.groupby(["pclass","sex"]).agg({"survived":["sum","count","mean"]})
```

```python
#########################################
# Görev 16:  30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazınız.
# Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz 
# (apply ve lambda yapılarını kullanınız)
#########################################

def age_30(age):
    if age<30:
        return 1
    else:
        return 0

df["age_flag"]=df["age"].apply(lambda x: age_30(x))
df.head()
```

```python
# veya 
df["age_flag"]=df["age"].apply(lambda x: 1 if x<30 else 0)
```

```python
#########################################
# Görev 17: Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.
#########################################

df=sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
#########################################
# Görev 18: Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill 
# değerlerinin toplamını, min, max ve ortalamasını bulunuz.
#########################################

df.groupby("time").agg({"total_bill":["sum","min","mean","max"]})
```

```python
#########################################
# Görev 19: Günlere ve time göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.
#########################################

df.groupby(["day","time"]).agg({"total_bill":["sum","min","max","mean"]})
```

```python
#########################################
# Görev 20:Lunch zamanına ve kadın müşterilere ait total_bill ve tip  değerlerinin 
# day'e göre toplamını, min, max ve ortalamasını bulunuz.
#########################################

df[(df["time"]=="Lunch") & (df["sex"]=="Female")].groupby("day").agg({"total_bill":["sum","min","max","mean"]})
```

```python
df[(df["time"] == "Lunch") & (df["sex"] == "Female")].groupby("day").agg({"total_bill": ["sum","min","max","mean"],})
```

```python
#########################################
# Görev 21: size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir?
#########################################

df.loc[(df["size"]<3) & (df["total_bill"]>10), "total_bill"].mean()
```

```python
#########################################
# Görev 22: total_bill_tip_sum adında yeni bir değişken oluşturun. 
# Her bir müşterinin ödediği totalbill ve tip in toplamını versin.
#########################################

df.head()
```

```python
df["total_bill_tip_sum"] = df["total_bill"]+df["tip"]
df.head()
```

```python
#########################################
# Görev 23: total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.
#########################################

new_df= df.sort_values("total_bill_tip_sum", ascending=False)
new_df.head(10)
```
