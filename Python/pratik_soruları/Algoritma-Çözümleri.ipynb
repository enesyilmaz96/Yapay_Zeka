# Coalesced

### Alıştırma_1

Coalesced yöntemi ile Hash Tablosunu oluşturunuz.

Hash(key) = key mod 11

    data=[27,18,29,28,39,13,16,42,17]


**Çalışma Mantığı:** 

Öncelikle 0 ile 10 arasında (11 indexli) bir Hash Tablosu oluşturulur. 

Verilerın sırasıyla Hash Tablosuna yerleştirilmesi gerekmektedir. Bunun için soruda verilen **`Key Mod 11`** kullanılır. Veri 11'e bölünür ve bölümünden kalan sayı, bize hash tablosundaki index numarasını verir. 

**ÖRN=** 27'nin 11'e bölümünden kalan 5'tir. Yani 27 verisi Hash tablsounun 5. indexine yerleştirilmesi gerekir.

Bu işlem çakışma olana kadar devam eder. Çakışma durumunda ise (Aynı index numarasına birden fazla veri yerleştirilmesi gerektiğinde) ilgili veri, en sondan başlayarak boş olan ilk indexe yerleştirilir.

**ÖRN=** 27 verisi 5. indexe ve 18 verisi 7. indexe yerleştirildi daha sonra 29 verisi geldi ve onun da 7. indexe yerleşmesi gerekiyor. Fakat Hash Tablosunda 7. indexe daha önce 18'i eklemiştik. Bu yüzden oraya yerleşemez. En sondan başlayarak boş index numaralarını kontrol ettiğimizde 10. index boş olduğundan 29 verisi 10. indexe yerleşir.


```python
def coalesced(data, hashTable):
    # Başlangıç indeksi
    i = 0
    # Veri dizisinin uzunluğu
    dongu = len(data)
    
    # Veri dizisindeki her eleman için döngü
    while i < dongu:
        # Veriyi uygun konuma yerleştirmek için hash fonksiyonu
        konum = data[0] % 11
        
        # Eğer hedef konum boşsa, veriyi yerleştir
        if hashTable[konum] == None:
            hashTable[konum] = data[0]
            # Veri dizisinden kullanılan veriyi çıkar
            data.pop(0)
            i += 1
        else:
            # Eğer hedef konum dolu ise çakışmayı çözmek için yeni bir konum belirle
            konum = 10
            while True:
                # Yeni konum boşsa, veriyi yerleştir ve döngüden çık
                if hashTable[konum] == None:
                    hashTable[konum] = data[0]
                    data.pop(0)
                    i += 1
                    break
                else:
                    # Eğer yeni konum da dolu ise bir önceki konuma git
                    konum -= 1

    return print(f"Hash Tablosu = {hashTable}")

# Test verileri
data = [27, 18, 29, 28, 39, 13, 16, 42, 17]
# 11 elemanlık bir hash tablosu oluştur
hashTable = [None] * 11
# Coalesced hashing işlemini uygula
coalesced(data, hashTable)
```

### Alıştırma_2

Coalesced yöntemi ile Hash Tablosunu oluşturunuz. Hash Tablosu oluştururken **`Pointer Tablosu`** kullanın.

Hash(key) = key mod 11

    data=[27,18,29,28,39,13,16,42,17] 
    
**Çalışma Mantığı:** 

Öncelikle 0 ile 10 arasında (11 indexli) bir Hash Tablosu oluşturulur. 

Verilerın sırasıyla Hash Tablosuna yerleştirilmesi gerekmektedir. Bunun için soruda verilen **`Key Mod 11`** kullanılır. Veri 11'e bölünür ve bölümünden kalan sayı, bize hash tablosundaki index numarasını verir. 

**ÖRN=** 27'nin 11'e bölümünden kalan 5'tir. Yani 27 verisi Hash tablsounun 5. indexine yerleştirilmesi gerekir.

Bu işlem çakışma olana kadar devam eder. Çakışma durumunda ise (Aynı index numarasına birden fazla veri yerleştirilmesi gerektiğinde) ilgili veri, en sondan başlayarak boş olan ilk indexe yerleştirilir. 

Pointer Tablosunun mantığı çakışan verinin kolay takip edilmesidir. Çakışma durumunda en sondaki boş hücreye yerleştirilen verinin, gerçekte yerleşmesi gerektiği index bellidir. Pointer Tablosunun bu index numarasına, çakışan verinin Hash Tablosunda hangi index numrasına yerleştirildiği yazılır. Böylece ilgili veri aranırken Hash Tablosunun indexine bakılır, eğer aranan veri orada yoksa Pointer Tablosuna bakarak Hash Tablosunun hangi indexine gitmesi gerektiğini bilir. Bu şekilde takibini gerçekleştirerek aradığı veriyi kolayca bulabilir.


**ÖRN=** 27 verisi 5. indexe ve 18 verisi 7. indexe yerleştirildi daha sonra 29 verisi geldi ve onun da 7. indexe yerleşmesi gerekiyor. Fakat Hash Tablosunda 7. indexe daha önce 18'i eklemiştik. Bu yüzden oraya yerleşemez. En sondan başlayarak boş index numaralarını kontrol ettiğimizde 10. index boş olduğundan 29 verisi 10. indexe yerleşir. Normalde 29'un Hash Tablosunda 7. indexe yerleştirilmesi gerekirken 10. indexe yerleştirdik. Bu sebeple Pointer Tablosunun 7. indexine 10 yazılır. Böylece ilgili veri aranırken öncelikle Hash Tablosunun 7. indexine bakılır eğer veri orada değilse Pointer Tablosunun 7. indexine bakılır. Orada yazan değer bize Hash Tablsounun kaçıncı indexine bakmamız gerektiğini söyler. Pointer Tablosundaki yönlendirme sayesinde 10. indexe bakılır ve aranan veri bulunur.

```python
def coalesced(data, hashTable, pointerTable):
    # Başlangıç indeksi
    i = 0
    # Veri dizisinin uzunluğu
    dongu = len(data)
    
    # Veri dizisindeki her eleman için döngü
    while i < dongu:
        # Veriyi uygun konuma yerleştirmek için hash fonksiyonu
        konum = data[0] % 11
        
        # Eğer hedef konum boşsa, veriyi yerleştir
        if hashTable[konum] == None:
            hashTable[konum] = data[0]
            # Veri dizisinden kullanılan veriyi çıkar
            data.pop(0)
            i += 1
        else:
            # Eğer hedef konum dolu ise çakışmaları çözmek için pointer tablosunu kullan
            pointerKonum = konum
            while True:
                # Pointer tablosunda boş bir konum bulana kadar ilerle
                if pointerTable[pointerKonum] != None:
                    pointerKonum = pointerTable[pointerKonum]
                else:
                    break
            
            # Yeni bir konum belirle
            konum = 10
            while True:                
                # Yeni konum boşsa, veriyi yerleştir ve pointer tablosunu güncelle
                if hashTable[konum] == None:
                    hashTable[konum] = data[0]
                    data.pop(0)
                    pointerTable[pointerKonum] = konum
                    i += 1
                    break
                else: 
                    # Eğer yeni konum da dolu ise bir önceki konuma git
                    konum -= 1
    
    # Sonuçları ekrana yazdır
    print(f"Hash Tablosu = {hashTable}\nPointer Tablosu = {pointerTable}")

# Test verileri
data = [27, 18, 29, 28, 39, 13, 16, 42, 17]
# 11 elemanlık bir hash tablosu ve pointer tablosu oluştur
hashTable = [None] * 11
pointerTable = [None] * 11
# Coalesced hashing işlemini uygula
coalesced(data, hashTable, pointerTable)
```

### Alıştırma_3

Coalesced yöntemi ile Hash Tablosunu oluşturunuz. Hash Tablosu oluştururken **`Pointer Tablosu`** kullanın. Algoritmayı tasarlarken  **`class`** kullanın.

Hash(key) = key mod 11

    data=[27,18,29,28,39,13,16,42,17] 

```python
class CoalescedHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size
        self.pointer_table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, item):
        home_index = self.hash_function(item)

        # Ana tablo boşsa, direkt ekleyebiliriz
        if self.table[home_index] is None:
            self.table[home_index] = item
        # Çakışma durumu: Pointer tabloya ekleyerek ilerle
        else:
            # Boş gücre aramaya en son indexden başlanır
            current_index = 10
            # Boş hücre bulana kadar index değerini bir azaltacak döngü oluşturduk.
            while self.table[current_index] is not None:                
                current_index -= 1
            self.table[current_index]=item
            
            # Pointer tablosunundaki index boşsa güncelledik           
            if self.pointer_table[home_index] is None:
                self.pointer_table[home_index]=current_index
            # Pointer tablosundaki index boş değilse boş hücreyi bulana kadar döngü oluşturduk.
            else:
                while self.pointer_table[home_index] is not None:
                    # Eğer Pointer tablosu dolu ise dolu hücredeki değerin olduğu indexe gidecek. Orası da doluysa
                    # bu sefer orda yazan indexe gidecek ta ki boş index değerini bulana kadar.
                    home_index=self.pointer_table[home_index]
                
                self.pointer_table[home_index]=current_index
    
    def display_tables(self):
        print("Hash Table:", self.table)
        print("Pointer Table:", self.pointer_table)

# Verilen örnek değerleri tabloya ekleyelim
items = [27, 18, 29, 28, 39, 13, 16, 42, 17]

# Coalesced Hashing tablosunu oluşturalım
hash_table = CoalescedHashTable(size=11)

for item in items:
    hash_table.insert(item)

# Tabloyu yazdıralım
hash_table.display_tables()
```

# Progressive Overflow

### Alıştırma_1

Progressive Overflow yöntemi ile Hash Tablosunu oluşturunuz.

Hash(key) = key mod 11

    data=[27,18,29,28,39,13,16]



**Çalışma Mantığı:** 

Öncelikle 0 ile 10 arasında (11 indexli) bir Hash Tablosu oluşturulur. 

Verilerın sırasıyla Hash Tablosuna yerleştirilmesi gerekmektedir. Bunun için soruda verilen **`Key Mod 11`** kullanılır. Veri 11'e bölünür ve bölümünden kalan sayı, bize hash tablosundaki index numarasını verir. 

**ÖRN=** 27'nin 11'e bölümünden kalan 5'tir. Yani 27 verisi Hash tablsounun 5. indexine yerleştirilmesi gerekir.

Bu işlem çakışma olana kadar devam eder. Çakışma durumunda ise (Aynı index numarasına birden fazla veri yerleştirilmesi gerektiğinde) index numarası birer birer arttırılır ve Hash Tablosunda boş bulduğu ilk yere yerleştirilir.

**ÖRN=** 27 verisi 5. indexe ve 18 verisi 7. indexe yerleştirildi daha sonra 29 verisi geldi ve onun da 7. indexe yerleşmesi gerekiyor. Fakat Hash Tablosunda 7. indexe daha önce 18'i eklemiştik. Bu yüzden oraya yerleşemez. İndex numarasını boş yer bulana kadar arttırmamız gerekiyor. İndexi bir arttırdığımızda 8. index boş olduğu için 29 verisi 8. indexe yerleşir.

**NOT :** İndex numarası arttırılırken son indexin 10 olduğunu unutmalayım. İndex numarası 10'a geldiğinde tekrar arttırmak istersek index numaramız 11 değil 0 olmalı. Yani index 0 ile 10 arasında kalmadı.



```python
def progressiveOverflow(data, hashTable):
    # Başlangıç indeksi
    i = 0
    # Veri dizisinin uzunluğu
    dongu = len(data)
    
    # Veri dizisindeki her eleman için döngü
    while i < dongu:
        # Veriyi uygun konuma yerleştirmek için hash fonksiyonu
        konum = data[0] % 11
        
        # Eğer hedef konum boşsa, veriyi yerleştir
        if hashTable[konum] == None:
            hashTable[konum] = data[0]
            # Veri dizisinden kullanılan veriyi çıkar
            data.pop(0)
            i += 1
        else:
            # Eğer hedef konum dolu ise ilerleyen konumları kontrol et
            while True:
                konum += 1
                # Konum 11'e ulaştığında, başa dön
                if konum == 11:
                    konum = 0
                # Boş bir konum bulunduğunda, veriyi yerleştir ve döngüden çık
                if hashTable[konum] == None:
                    hashTable[konum] = data[0]
                    data.pop(0)
                    i += 1
                    break
                    
    return print(f"Hash Tablosu = {hashTable}")

# Test verileri
data = [27, 18, 29, 28, 39, 13, 16]
# 11 elemanlık bir hash tablosu oluştur
hashTable = [None] * 11
# Progressive Overflow hashing işlemini uygula
progressiveOverflow(data, hashTable)
```

### Alıştırma_2

Progressive Overflow yöntemi ile "Alıştırma_1" de olduğu gibi Hash Tablosunu oluşturunuz. Algoritmayı tasarlarken **`enumerate`** kullanarak kodu kısaltınız.


Hash(key) = key mod 11

    data=[27,18,29,28,39,13,16,42,17] 

```python
def progressive_overflow(data, hash_table):
    for index, veri in enumerate(data):
        konum = veri % 11  # Verinin hash tablosundaki konumu hesapla
        
        # Eğer konumda değer yoksa, direkt olarak veriyi ekleyebiliriz
        if hash_table[konum] == None:
            hash_table[konum] = veri
        else:
            # Çakışma durumu: Bir sonraki boş konumu bulana kadar ilerle
            while hash_table[konum] != None:
                konum += 1
            
            # Boş konumu bulduktan sonra veriyi ekleyebiliriz
            hash_table[konum] = veri
    
    return print(f"Hash Tablosu = {hash_table}")

# Örnek veri ve hash table
data = [27, 18, 29, 28, 39, 13, 16]
hash_table = [None] * 11

# Progressive Overflow Hashing uygula
progressive_overflow(data, hash_table)
```

### Alıştırma_3

Progressive Overflow yöntemi ile Hash Tablosu'nu **`Linked List`** kullanarak oluşturunuz.

Hash(key) = key mod 11

    data=[27,18,29,28,39,13,16,5]


**Çalışma Mantığı:** 

Yukarıdaki işlemden farklı olarak çakışma olması durumunda Linked List'e bakılır. Eğer Linked List'in ilgili indexi boşsa, veri oraya yerleştirilir. Boş değilse bir önceki yöntemde olduğu gibi Hash Tablosunda index numarası birer birer arttırılarak boş bulunun ilk yere yerleştirilir.

**ÖRN=** 27 verisi 5. indexe ve 18 verisi 7. indexe yerleştirildi daha sonra 29 verisi geldi ve onun da 7. indexe yerleşmesi gerekiyor. Fakat Hash Tablosunda 7. indexe daha önce 18'i eklemiştik. Bu yüzden oraya yerleşemez ve bu durumda Linked List'in 7. indexi kontrol edilir. Linked List'in 7. indexi boş olduğu için 29 verisi Linked List'in 7. indexine yerleştirilir.

**Çakışma durumunda Linked List indexi de boş olmasaydı;**

Yukarıdaki örneğe ek olarak "5" verisini ekleyelim. 5'in 11'e bölümünden kalan 5'tir. Fakat hem Hash Tablosunda hemde Linked Listte 5. indexe daha önce veri yerleştirdik. Bu sebeple Hash Tablosunda boş yer bulana kadar index numarası bir arttılır. İlk boş bulunan index numrasına (8. index) yerleştirilir.


```python
def progressiveOverflow(data, hashTable, linkedList):
    # Başlangıç indeksi
    i = 0
    # Veri dizisinin uzunluğu
    dongu = len(data)
    
    # Veri dizisindeki her eleman için döngü
    while i < dongu:
        # Veriyi uygun konuma yerleştirmek için hash fonksiyonu
        konum = data[0] % 11
        
        # Eğer hedef konum boşsa, veriyi hash tablosuna yerleştir
        if hashTable[konum] == None:
            hashTable[konum] = data[0]
            # Veri dizisinden kullanılan veriyi çıkar
            data.pop(0)
            i += 1
        # Eğer hedef konum dolu ve linkedList boşsa, veriyi linkedList'e yerleştir
        elif linkedList[konum] == None:
            linkedList[konum] = data[0]
            # Veri dizisinden kullanılan veriyi çıkar
            data.pop(0)
            i += 1
        else:
            # Eğer hedef konum dolu ve linkedList de dolu ise ilerleyen konumları kontrol et
            while True:
                konum += 1
                # Konum 11'e ulaştığında, başa dön
                if konum == 11:
                    konum = 0
                # Boş bir konum bulunduğunda, veriyi hash tablosuna yerleştir ve döngüden çık
                if hashTable[konum] == None:
                    hashTable[konum] = data[0]
                    # Veri dizisinden kullanılan veriyi çıkar
                    data.pop(0)
                    i += 1
                    break
                                                          
    # Sonuçları ekrana yazdır
    print(f"Hash Tablosu = {hashTable}\nLinked List = {linkedList}")

# Test verileri
data = [27, 18, 29, 28, 39, 13, 16, 5]
# 11 elemanlık bir hash tablosu ve linked list oluştur
hashTable = [None] * 11
linkedList = [None] * 11
# Progressive Overflow hashing işlemini uygula
progressiveOverflow(data, hashTable, linkedList)
```

### Alıştırma_4

Progressive Overflow yöntemi ile Hash Tablosu'nu **`Linked List`** kullanarak oluşturunuz. Algoritmayı tasarlarken **`class`** kullanınız.

Hash(key) = key mod 11

    data=[27,18,29,28,39,13,16,5]

```python
class ProgressiveOverflowHashTable:
    def __init__(self, size):
        # Tablonun boyutunu ve ana tablo ile gelişen tabloyu oluştur
        self.size = size
        self.home_table = [None] * size
        self.overflow_table = [None] * size

    def hash_function(self, key):
        # Anahtarın tabloya yerleştirileceği indeksi belirle
        return key % self.size

    def insert(self, item):
        # Anahtarın ana tabloya yerleştirileceği indeksi hesapla
        home_index = self.hash_function(item)

        # Ana tablo boşsa, direkt ekleyebiliriz
        if self.home_table[home_index] is None:
            self.home_table[home_index] = item
        else:
            # Çakışma durumu: Gelişen tabloya ekle
            overflow_index = self.hash_function(home_index)
            while self.overflow_table[overflow_index] is not None:
                # Boş bir konum bulana kadar ilerle
                overflow_index = self.hash_function(overflow_index + 1)
            self.overflow_table[overflow_index] = item

    def display_tables(self):
        # Ana tablo ve gelişen tabloyu ekrana yazdır
        print("Hash Tablo:", self.home_table)
        print("Linked List:", self.overflow_table)


# Verilen örnek değerleri tabloya ekleyelim
items = [27, 18, 29, 28, 39, 13, 16,5]

# Progressive Overflow tablosunu oluşturalım
hash_table = ProgressiveOverflowHashTable(size=11)

for item in items:
    hash_table.insert(item)

# Tabloyu yazdıralım
hash_table.display_tables()
```

### Alıştırma_5

Progressive Overflow yöntemi ile Hash Tablosu'nu **`Linked List`** kullanarak oluşturunuz. Algoritmayı tasarlarken **`enumerate`** kullanarak kodu kısaltınız.

Hash(key) = key mod 11

    data=[27,18,29,28,39,13,16,5]

```python
def progressive_overflow(data, hash_table, linked_list):
    for index, veri in enumerate(data):
        konum = veri % 11  # Verinin hash tablosundaki konumu hesapla
        
        # Eğer konumda değer yoksa, direkt olarak veriyi hash table'a ekleyebiliriz
        if hash_table[konum] == None:
            hash_table[konum] = veri
        
        # Eğer konumda hash table'da değer varsa, linked list kullanarak çakışmayı çöz
        elif linked_list[konum] == None:
            linked_list[konum] = veri
            
        else:
            # Çakışma durumu: Bir sonraki boş konumu bulana kadar ilerle
            while hash_table[konum] != None:
                konum += 1
            
            # Boş konumu bulduktan sonra veriyi hash table'a ekleyebiliriz
            hash_table[konum] = veri
    
    return print(f'Hash Tablosu = {hash_table} \nLinked List = {linked_list}')

# Örnek veri, hash table ve linked list
data = [27, 18, 29, 28, 39, 13, 16, 5]
hash_table = [None] * 11
linked_list = [None] * 11

# Progressive Overflow Hashing uygula
progressive_overflow(data, hash_table, linked_list)
```

# Linear Quatient 

### Alıştırma_1

Linear Quatient yöntemi ile Hash Tablosunu oluşturunuz.

Hash(key) = key mod 11

data=[27,18,29,28,39,13,16]

**Çalışma Mantığı:** 

Öncelikle 0 ile 10 arasında (11 indexli) bir Hash Tablosu oluşturulur. 

Verilerın sırasıyla Hash Tablosuna yerleştirilmesi gerekmektedir. Bunun için soruda verilen **`Key Mod 11`** kullanılır. Veri 11'e bölünür ve bölümünden kalan sayı, bize hash tablosundaki index numarasını verir. 

**ÖRN=** 27'nin 11'e bölümünden kalan 5'tir. Yani 27 verisi Hash tablsounun 5. indexine yerleştirilmesi gerekir.

Bu işlem çakışma olana kadar devam eder. Çakışma durumunda ise (Aynı index numarasına birden fazla veri yerleştirilmesi gerektiğinde) index numarası bellirli bir oranda arttırılarak Hash Tablosunda boş bulduğu ilk yere yerleştirilir. 

İndex numrasının ne kadar arttırılacağı ise verinin 11'e bölümünden elde edilen sonuç ile belirlenir. Hash(key), bölme işleminden kalan ile belirleniyordu. Buradaki fark ise bölme işleminin sonucundaki tam sayıdan elde edilmesidir. Yani 27/11=2,45 yapar. Fakat tam sayı almamız gerektiğinden buradan 2 sonucu gelir. 27 verisi için çakışma olması durumunda index ikişer ikişer arttırılır.

**ÖRN=** 27 verisi 5. indexe ve 18 verisi 7. indexe yerleştirildi daha sonra 29 verisi geldi ve onun da 7. indexe yerleşmesi gerekiyor. Fakat Hash Tablosunda 7. indexe daha önce 18'i eklemiştik. Bu yüzden oraya yerleşemez. 29/11=2,63 yapar tam sayı olarak 2 geldiğinden İndex numarasını boş yer bulana kadar ikişer ikişier arttırmamız gerekiyor. İndexi iki arttırdığımızda 9. index boş olduğu için 29 verisi 9. indexe yerleşir.

**NOT :** İndex numarası arttırılırken son indexin 10 olduğunu unutmalayım. İndex numarası 10'a geldiğinde tekrar arttırmak istersek index numaramız 11 değil 0 olmalı. Yani index 0 ile 10 arasında kalmadı.


```python
def linearQuatient(data, hashTable):
    # Başlangıç indeksi
    i = 0
    # Veri dizisinin uzunluğu
    dongu = len(data)
    
    # Veri dizisindeki her eleman için döngü
    while i < dongu:
        # Veriyi uygun konuma yerleştirmek için hash fonksiyonu
        konum = data[0] % 11
        
        # Eğer hedef konum boşsa, veriyi yerleştir
        if hashTable[konum] == None:
            hashTable[konum] = data[0]
            # Veri dizisinden kullanılan veriyi çıkar
            data.pop(0)
            i += 1
        else:
            # Eğer hedef konum dolu ise, lineer quatient kullanarak bir sonraki boş konumu bul
            while True:
                konum += data[0] // 11
                # Konum 10'u geçerse başa dön
                if konum > 10:
                    konum -= 11
                # Boş bir konum bulunduğunda, veriyi yerleştir ve döngüden çık
                if hashTable[konum] == None:
                    hashTable[konum] = data[0]
                    # Veri dizisinden kullanılan veriyi çıkar
                    data.pop(0)
                    i += 1
                    break
                    
    return print(f"Hash Tablosu = {hashTable}")

# Test verileri
data = [27, 18, 29, 28, 39, 13, 16]
# 11 elemanlık bir hash tablosu oluştur
hashTable = [None] * 11
# Linear Quatient hashing işlemini uygula
linearQuatient(data, hashTable)
```

### Alıştırma_3

Hash(key) = key mod 11

```python
class LinearQuotientHashTable:
    def __init__(self, size):
        # Tablonun boyutunu ve ana tabloyu oluştur
        self.size = size
        self.table = [None] * size

    def hash_function1(self, key):
        # Anahtarın tabloya yerleştirileceği indeksi belirle
        return key % self.size

    def hash_function2(self, key):
        # İkinci bir hash fonksiyonu kullanarak lineer quatient belirle
        return (key // self.size) % self.size

    def insert(self, item):
        # Anahtarın ana tabloya yerleştirileceği indeksi hesapla
        home_index = self.hash_function1(item)

        # Ana tablo boşsa, direkt ekleyebiliriz
        if self.table[home_index] is None:
            self.table[home_index] = item
        else:
            i = 1
            while True:
                # Çakışma durumu: Linear Quotient ile uygun konuma ilerle
                current_index = (home_index + i * self.hash_function2(item)) % self.size
                if self.table[current_index] is None:
                    self.table[current_index] = item
                    break
                i += 1

    def display_table(self):
        # Ana tabloyu ekrana yazdır
        print("Ana Tablo:", self.table)


# Verilen örnek değerleri tabloya ekleyelim
items = [27, 18, 29, 28, 39, 13, 16]

# Linear Quotient Hashing tablosunu oluşturalım
hash_table = LinearQuotientHashTable(size=11)

for item in items:
    hash_table.insert(item)

# Tabloyu yazdıralım
hash_table.display_table()
```

```python
class Node:
    def __init__(self, data):
        # Düğümün sol ve sağ alt ağaçları ve verisi
        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
        # Yeni değeri üst düğümle karşılaştırın
        if self.data:
            if data < self.data:
                # Sol alt ağaçta yeni düğüm oluşturun veya var olan alt ağaca gidin
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                # Sağ alt ağaçta yeni düğüm oluşturun veya var olan alt ağaca gidin
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            # Düğüm verisi boşsa, veriyi bu düğüme atayın
            self.data = data

    def PrintTree(self):
        # Ağacı inorder (sol, kök, sağ) sırasında yazdırın
        if self.left:
            self.left.PrintTree()
        print(self.data),
        if self.right:
            self.right.PrintTree()

# Düğüm eklemek için insert yöntemini kullanın
root = Node(12)
root.insert(6)
root.insert(14)
root.insert(3)

# Ağacı yazdır
root.PrintTree()
```

# Binary Tree Hash Table

### Alıştırma_1

Binary Tree ve  yöntemi ile Hash Tablosunu oluşturunuz. Algoritmayı tasarlarken **`class`** kullanın.

Hash(key) = key mod 11

    data= [27, 18, 29, 28, 39, 13]

```python
class BinaryTreeHashTable:
    def __init__(self, size):
        # Hash tablosunun boyutunu belirleyin
        self.size = size
        # Hash tablosunu oluşturun, başlangıçta her elemanı None olarak ayarlayın
        self.table = [None] * size

    def hash_function1(self, key):
        # Hash fonksiyonu 1: Mod alma işlemi ile anahtarın tabloya yerleştirileceği indeksi belirle
        return key % self.size

    def hash_function2(self, key):
        # Hash fonksiyonu 2: İkinci bir hash fonksiyonu kullanarak lineer quatient belirle
        return (key // self.size) % self.size

    def continue_function(self, node, item):
        # Çakışma durumu: Lineer quatient ile uygun konuma ilerle
        i = self.hash_function2(item)
        left = node + i
        # Konum 10'u geçerse başa dön
        if left > 10:
            left -= 11
        return left

    def move_function(self, node):
        # Çakışma durumu devam ederse: İkinci bir konum belirle
        right = node + (self.table[node] // 11)
        # Konum 10'u geçerse başa dön
        if right > 10:
            right -= 11
        return right

    def insert(self, item):
        # Anahtarın tabloya yerleştirileceği indeksi hash fonksiyonu 1 ile belirle
        home_index = self.hash_function1(item)

        # Ana tablo boşsa, direkt ekleyebiliriz
        if self.table[home_index] is None:
            self.table[home_index] = item
        else:
            # Çakışma durumu: Lineer quatient ile uygun konuma ilerle
            left = self.continue_function(home_index, item)

            # Eğer uygun konum boşsa, veriyi ekleyin
            if self.table[left] is None:
                self.table[left] = item
            else:
                # Çakışma durumu devam ederse: İkinci bir konum belirle
                right = self.move_function(home_index)
                
                # Eğer ikinci konum boşsa, veriyi ekleyin ve yer değiştirin
                if self.table[right] is None:
                    self.table[right] = self.table[home_index]
                    self.table[home_index] = item
                else:
                    # İkinci konum da doluysa, uygun boş konumu bulana kadar devam et
                    while True:
                        left = self.continue_function(home_index, left)
                        if self.table[left] is None:
                            self.table[left] = item
                            break
                        right = self.move_function(home_index)
                        if self.table[right] is None:
                            self.table[right] = self.table[home_index]
                            self.table[home_index] = item
                            break

    def display_table(self):
        # Hash tablosunu ekrana yazdırın
        print("Ana Tablo:", self.table)


# Verilen örnek değerleri tabloya ekleyelim
items = [27, 18, 29, 28, 39, 13]

# BinaryTreeHashTable sınıfını kullanarak tabloyu oluşturalım
hash_table = BinaryTreeHashTable(size=11)

for item in items:
    hash_table.insert(item)

# Tabloyu yazdıralım
hash_table.display_table()
```

# Binary Search Tree - BST

Bu Python kodu, bir ikili arama ağacı (Binary Search Tree - BST) oluşturan ve bu ağaç üzerinde sıralı geçiş yapabilen bir sınıf ve yardımcı fonksiyonları içerir.

1. `Node` sınıfı, BST'nin düğümlerini temsil eder. Her düğüm, bir sol alt ağacı, bir sağ alt ağacı ve bir değeri (val) içerir.

2. `insert` fonksiyonu, verilen bir anahtar değeriyle yeni bir düğüm ekler. Eğer ağaç boşsa, yeni bir kök düğüm oluşturulur. Eğer ağaç boş değilse, anahtarın değerine göre uygun alt ağaca eklenir. Eğer anahtar zaten ağaçta varsa, aynı düğümü döndürür (her anahtarın eşsiz olması gerektiği varsayımıyla).

3. `inorder` fonksiyonu, BST üzerinde sıralı geçişi gerçekleştirir. Sol alt ağacı, düğümün değeri, ve sağ alt ağacı sırasıyla ziyaret edilir. Bu, BST'nin "in-order traversal" özelliğini kullanarak düğümleri küçükten büyüğe doğru sıralı bir şekilde yazdırır.

Sürücü programı ise `Node` sınıfını kullanarak bir BST oluşturur ve ardından oluşturulan BST'nin sıralı geçişini `inorder` fonksiyonu aracılığıyla ekrana yazdırır. Örnek bir BST oluşturulmuş ve sıralı geçişi yapılmıştır.

```python
# Temsil eden bir yardımcı sınıf
# BST'deki tek bir düğüm
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

# Eklenecek bir yardımcı fonksiyon
# Verilen anahtarla yeni bir düğüm oluşturur ve BST'ye ekler
def insert(root, key):
    if root is None:
        return Node(key)
    else:
        # Eğer kök düğüm None değilse, anahtarın uygun konumunu bulana kadar rekürsif olarak ilerler
        if root.val == key:
            return root
        elif root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

# Sıralı ağaç geçişi yapmak için bir yardımcı fonksiyon
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

# Yukarıdaki fonksiyonları test etmek için sürücü programı
if __name__ == '__main__':
    # Aşağıdaki BST'yi oluşturalım
    #       27
    #      /  \
    #     18   29
    #    / \   / \
    #   13  19 28 39
    #       /      \
    #      16       41

    r = Node(27)
    r = insert(r, 18)
    r = insert(r, 29)
    r = insert(r, 28)
    r = insert(r, 39)
    r = insert(r, 13)
    r = insert(r, 16)
    r = insert(r, 41)
    r = insert(r, 17)
    r = insert(r, 19)

    # BST'nin sıralı geçişini yazdırma
    inorder(r)
```

# Water Jug Problem:

**(Kaynak = Artificial Intelligence, sayfa: 28)**

  Su damlası problemi (Water Jug Problem) genellikle su miktarlarını kontrol etme ve belirli bir hedef su miktarına ulaşma problemi olarak formüle edilir. Burada, iki adet su kabı düşünülmekte ve bu kabların kapasiteleri belirlenmiştir. Hedef, belirli bir su miktarını belirli bir kabın içine doldurmaktır.

Aşağıda, bu problemin çözümü için Python dilinde bir kod örneği bulunmaktadır:

```python
def water_jug_problem(capacity_x, capacity_y, target):
    state = (0, 0)  # Başlangıç durumu (su miktarları)
    visited_states = set()  # Ziyaret edilen durumları takip etmek için küme

    while state != target:
        visited_states.add(state)

        x, y = state
        # 1. Durum: X kabını doldur
        fill_x = (capacity_x, y)
        if fill_x not in visited_states:
            print(f"Fill X: {fill_x}")
            state = fill_x
            continue

        # 2. Durum: Y kabını doldur
        fill_y = (x, capacity_y)
        if fill_y not in visited_states:
            print(f"Fill Y: {fill_y}")
            state = fill_y
            continue

        # 3. Durum: X kabından Y kabına su aktar
        pour_x_to_y = (max(0, x - (capacity_y - y)), min(capacity_y, y + x))
        if pour_x_to_y not in visited_states:
            print(f"Pour X to Y: {pour_x_to_y}")
            state = pour_x_to_y
            continue

        # 4. Durum: Y kabından X kabına su aktar
        pour_y_to_x = (min(capacity_x, x + y), max(0, y - (capacity_x - x)))
        if pour_y_to_x not in visited_states:
            print(f"Pour Y to X: {pour_y_to_x}")
            state = pour_y_to_x
            continue

        # 5. Durum: X kabını boşalt
        empty_x = (0, y)
        if empty_x not in visited_states:
            print(f"Empty X: {empty_x}")
            state = empty_x
            continue

        # 6. Durum: Y kabını boşalt
        empty_y = (x, 0)
        if empty_y not in visited_states:
            print(f"Empty Y: {empty_y}")
            state = empty_y
            continue

    print("Hedefe ulaşıldı!")

# Örnek kullanım
water_jug_problem(4, 3, (2, 0))

```

# Breadth First Search

```python
from collections import deque

def water_jug_bfs(capacity_x, capacity_y, target):
    initial_state = (0, 0)
    visited_states = set()
    queue = deque([(initial_state, [])])

    while queue:
        current_state, path = queue.popleft()
        visited_states.add(current_state)

        if current_state == target:
            print("Hedefe ulaşıldı!")
            print("Yol:", path)
            return

        x, y = current_state

        # 1. Durum: X kabını doldur
        fill_x = (capacity_x, y)
        if fill_x not in visited_states:
            queue.append((fill_x, path + ["Fill X"]))

        # 2. Durum: Y kabını doldur
        fill_y = (x, capacity_y)
        if fill_y not in visited_states:
            queue.append((fill_y, path + ["Fill Y"]))

        # 3. Durum: X kabından Y kabına su aktar
        pour_x_to_y = (max(0, x - (capacity_y - y)), min(capacity_y, y + x))
        if pour_x_to_y not in visited_states:
            queue.append((pour_x_to_y, path + ["Pour X to Y"]))

        # 4. Durum: Y kabından X kabına su aktar
        pour_y_to_x = (min(capacity_x, x + y), max(0, y - (capacity_x - x)))
        if pour_y_to_x not in visited_states:
            queue.append((pour_y_to_x, path + ["Pour Y to X"]))

        # 5. Durum: X kabını boşalt
        empty_x = (0, y)
        if empty_x not in visited_states:
            queue.append((empty_x, path + ["Empty X"]))

        # 6. Durum: Y kabını boşalt
        empty_y = (x, 0)
        if empty_y not in visited_states:
            queue.append((empty_y, path + ["Empty Y"]))

# Örnek kullanım
water_jug_bfs(4, 3, (2, 0))
```

Her hamle sonrasında oyun durumu kontrol edilir ve bir oyuncunun kazanıp kazanmadığı veya berabere olup olmadığı kontrol edilir. Oyunun tahtası ve durumu sürekli olarak ekrana basılır. Aşağıdaki oyun iki kişilik versiyonudur.

```python
# Oyun tahtasını ekrana basan fonksiyon
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

# Oyun tahtasında bir oyuncunun kazanıp kazanmadığını kontrol eden fonksiyon
def check_winner(board, player):
    # Satırları kontrol et
    for row in board:
        if all(cell == player for cell in row):
            return True

    # Sütunları kontrol et
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True

    # Köşegenleri kontrol et
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

# Oyun tahtasının dolup dolmadığını kontrol eden fonksiyon
def is_board_full(board):
    return all(cell != " " for row in board for cell in row)

# X ve O oyuncuları arasında oynanan Tic-Tac-Toe oyununu yöneten fonksiyon
def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]  # Boş bir oyun tahtası oluştur
    current_player = "X"  # Başlangıçta X oyuncusu hamle yapar

    while True:
        print_board(board)  # Oyun tahtasını ekrana bas

        # Kullanıcıdan hamle al
        row = int(input(f"Player {current_player}, enter row (0-2): "))
        col = int(input(f"Player {current_player}, enter column (0-2): "))

        # Girilen hücre zaten doluysa uyarı ver ve tekrar hamle al
        if board[row][col] != " ":
            print("Cell already taken. Try again.")
            continue

        board[row][col] = current_player  # Oyuncunun hamlesini tahtaya işaretle

        # Oyuncunun kazanıp kazanmadığını kontrol et
        if check_winner(board, current_player):
            print_board(board)
            print(f"Player {current_player} wins!")
            break

        # Oyun tahtası doluysa ve kazanan yoksa berabere (tie) ilan et
        if is_board_full(board):
            print_board(board)
            print("It's a tie!")
            break

        # Sıradaki oyuncuya geç
        current_player = "O" if current_player == "X" else "X"

# Ana programı başlat
if __name__ == "__main__":
    tic_tac_toe()
```

## Tic-Tac-Toe vs. Computer

Bu oyun, oyuncunun ve bilgisayarın sırayla hamleler yapmasına izin verir ve kazananı veya beraberliği kontrol eder. Oyun tahtası her hamle sonrasında ekrana basılır ve oyuncudan veya bilgisayardan hamle alınır. Oyunun sonunda kazanan veya beraberlik durumu ekrana basılır. 

```python
import random

# Oyun tahtasını ekrana basan fonksiyon
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

# Oyun tahtasında bir oyuncunun kazanıp kazanmadığını kontrol eden fonksiyon
def check_winner(board, player):
    # Satırları kontrol et
    for row in board:
        if all(cell == player for cell in row):
            return True

    # Kolonları kontrol et
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True

    # Köşegenleri kontrol et
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

# Oyun tahtasının dolup dolmadığını kontrol eden fonksiyon
def is_board_full(board):
    return all(cell != " " for row in board for cell in row)

# Oyuncudan hamle almak için fonksiyon
def get_player_move(board, player):
    while True:
        row = int(input(f"Player {player}, enter row (0-2): "))
        col = int(input(f"Player {player}, enter column (0-2): "))

        if board[row][col] == " ":
            return row, col
        else:
            print("Cell already taken. Try again.")

# Bilgisayarın hamlesini belirlemek için fonksiyon
def get_computer_move(board):
    while True:
        row = random.randint(0, 2)
        col = random.randint(0, 2)

        if board[row][col] == " ":
            return row, col

# Oyunu yöneten ana fonksiyon
def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]  # Boş bir oyun tahtası oluştur
    current_player = "X"  # Başlangıçta X oyuncusu hamle yapar

    while True:
        print_board(board)  # Oyun tahtasını ekrana bas

        if current_player == "X":
            row, col = get_player_move(board, current_player)  # Oyuncudan hamle al
        else:
            print("Computer is making a move...")
            row, col = get_computer_move(board)  # Bilgisayarın hamlesini belirle

        board[row][col] = current_player  # Hamleyi tahtaya işaretle

        # Oyuncunun veya bilgisayarın kazanıp kazanmadığını kontrol et
        if check_winner(board, current_player):
            print_board(board)
            if current_player == "X":
                print("Player X wins!")
            else:
                print("Computer wins!")
            break

        # Oyun tahtası doluysa ve kazanan yoksa berabere (tie) ilan et
        if is_board_full(board):
            print_board(board)
            print("It's a tie!")
            break

        current_player = "O" if current_player == "X" else "X"  # Sıradaki oyuncuya geç

if __name__ == "__main__":
    tic_tac_toe()
```
